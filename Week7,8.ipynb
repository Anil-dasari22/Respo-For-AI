```python
# (a) Not all cars have carburetors.
cars = {"car1", "car2"}
cars_with_carburetor = {"car1"}
def lacks_carburetor():
    return [c for c in cars if c not in cars_with_carburetor]

# (b) Some people are either religious or pious.
people = {"alice", "bob"}
religious = {"alice"}
pious = {"bob"}
def religious_or_pious():
    return [p for p in people if p in religious or p in pious]

# (c) No dogs are intelligent.
dogs = {"dog1", "dog2"}
intelligent = {"human1", "human2"}
def is_dog_intelligent():
    return any(d in intelligent for d in dogs)  # should be False

# (d) All babies are illogical.
babies = {"tommy", "emma"}
def is_illogical(person):
    return person in babies

# (e) Every number is either negative or has a square root.
numbers = [-3, -1, 0, 1, 4]
def has_property(num):
    return num < 0 or num >= 0  # has square root if >=0

# (f) Some numbers are not real.
numbers_not_real = {"i"}  # imaginary number
def nonreal_numbers():
    return numbers_not_real

# (g) Every connected and circuit-free graph is a tree.
graphs = {"g1", "g2"}
connected = {"g1"}
circuit_free = {"g1"}
def is_tree(g):
    return g in connected and g in circuit_free

# (a) Not every graph is connected.
def not_all_graphs_connected():
    return [g for g in graphs if g not in connected]

# (b) All that glitters is not gold.
glitters = {"fake_jewel", "plastic_ring", "gold_ring"}
gold = {"gold_ring"}
def all_that_glitters_not_gold():
    return [x for x in glitters if x not in gold]

# (c) Not all that glitters is gold.
def some_glitters_not_gold():
    return [x for x in glitters if x not in gold]

# (d) The barber who shaves all men who do not shave themselves.
men = {"john", "david"}
in_town = {"john", "david"}
shaves_self = {"john"}  # only john shaves himself
barber = "bob"

def shaves(barber_name, person):
    return barber_name == barber and person in men and person in in_town and person not in shaves_self

# (e) There is no business-like show business.
show_business = {"show1", "show2"}
business_like = set()  # none are business-like
def not_business_like():
    return [x for x in show_business if x not in business_like]

# ---------------------------------------
# Run Demonstrations
# ---------------------------------------
if __name__ == "__main__":
    print("(a) Cars without carburetor:", lacks_carburetor())
    print("(b) People who are religious or pious:", religious_or_pious())
    print("(c) Any intelligent dogs?:", is_dog_intelligent())
    print("(d) Babies are illogical:", [b for b in babies if is_illogical(b)])
    print("(e) Numbers and property:", [(n, has_property(n)) for n in numbers])
    print("(f) Non-real numbers:", nonreal_numbers())
    print("(g) Trees:", [g for g in graphs if is_tree(g)])
```

```python
# Import logic functions from Python standard library
import itertools

# Define logical implication function
def implies(p, q):
    return (not p) or q

# Define the first logical problem (Neetha example)
def check_neetha_logic():
    """
    Premises:
    1. Every CS student takes Discrete Mathematics.      -> (CS(x) → DM(x))
    2. Neetha takes Discrete Mathematics.                -> DM(Neetha)
    Conclusion:
    Therefore, Neetha is a CS student.                   -> CS(Neetha)
    This conclusion is FALSE (invalid inference).
    """

    print("=== Checking Logic: Neetha and Discrete Mathematics ===")
    print("Premises:")
    print("1. Every CS student takes Discrete Mathematics. (CS → DM)")
    print("2. Neetha takes Discrete Mathematics. (DM)")
    print("Conclusion: Therefore, Neetha is a CS student. (CS)\n")

    # For all possible truth values of CS and DM
    for cs, dm in itertools.product([True, False], repeat=2):
        # Premise 1: CS → DM
        premise1 = implies(cs, dm)
        # Premise 2: DM is True (Neetha takes DM)
        premise2 = dm
        # Conclusion: CS (Neetha is CS student)
        conclusion = cs

        # Check validity: If all premises true, then conclusion must also be true
        valid = implies(premise1 and premise2, conclusion)

        print(f"CS={cs}, DM={dm} -> Premises: {premise1 and premise2}, Conclusion: {conclusion}, Valid: {valid}")

    print("\nResult: The conclusion 'Neetha is a CS student' is NOT logically valid (False inference).\n")


# Define the second logical problem (Rain, Fog, Trophy example)
def check_rain_logic():
    """
    Statements:
    1. If it does not rain or it is not foggy → sailing race & lifesaving demos go on.
    2. If sailing race is held → trophy awarded.
    3. Trophy was not awarded.
    Conclusion: Therefore, it rained.
    """

    print("=== Checking Logic: Weather and Sailing Race Example ===")
    print("Premises:")
    print("1. (¬Rain ∨ ¬Fog) → (Sailing ∧ Demo)")
    print("2. Sailing → Trophy")
    print("3. ¬Trophy")
    print("Conclusion: Therefore, Rain is TRUE.\n")

    # Generate all possible truth values
    for rain, fog, sailing, demo, trophy in itertools.product([True, False], repeat=5):
        premise1 = implies((not rain) or (not fog), sailing and demo)
        premise2 = implies(sailing, trophy)
        premise3 = not trophy

        # All premises must be True
        all_premises = premise1 and premise2 and premise3

        # Conclusion: It rained
        conclusion = rain

        # Check if argument valid: premises → conclusion
        valid = implies(all_premises, conclusion)

        if all_premises:
            print(f"Rain={rain}, Fog={fog}, Sailing={sailing}, Demo={demo}, Trophy={trophy} -> Conclusion (Rain)={conclusion}, Valid={valid}")

    print("\nResult: Based on premises, it logically follows that it must have RAINED.\n")


# Main function
def main():
    print("Logic Reasoning Verification using Python\n")
    check_neetha_logic()
    check_rain_logic()

# Run the program
if __name__ == "__main__":
    main()
```

```python
import itertools

# Logical implication function
def implies(p, q):
    return (not p) or q

def check_doctor_golfer_logic():
    print("=== Checking Logic: Doctors, Golfers, and College Graduates ===")
    print("Premises:")
    print("1. All doctors are college graduates. (∀x: D(x) → C(x))")
    print("2. Some doctors are not golfers. (∃x: D(x) ∧ ¬G(x))")
    print("Conclusion: Some golfers are not college graduates. (∃x: G(x) ∧ ¬C(x))\n")

    # Universe of people
    people = ['A', 'B', 'C']

    # Generate all possible truth assignments for D, G, C for each person
    # Each person can either be a doctor, golfer, or college graduate (True/False)
    for assignment in itertools.product([True, False], repeat=9):
        D = dict(zip(people, assignment[0:3]))  # Doctors
        G = dict(zip(people, assignment[3:6]))  # Golfers
        C = dict(zip(people, assignment[6:9]))  # College grads

        # Premise 1: For all x, D(x) → C(x)
        premise1 = all(implies(D[p], C[p]) for p in people)
        # Premise 2: There exists x such that D(x) ∧ ¬G(x)
        premise2 = any(D[p] and not G[p] for p in people)
        # Conclusion: There exists x such that G(x) ∧ ¬C(x)
        conclusion = any(G[p] and not C[p] for p in people)

        # Check if premises true but conclusion false (counterexample)
        if premise1 and premise2 and not conclusion:
            print("❌ Counterexample found!")
            print("Doctors:", D)
            print("Golfers:", G)
            print("College Graduates:", C)
            print("Premises True, Conclusion False — Argument INVALID.\n")
            return  # Stop at first invalid case

    print("✅ No counterexample found. Argument valid (but this won’t happen here).")

if __name__ == "__main__":
    check_doctor_golfer_logic()
```
